```python
import graphlab
```

```python
train_data = graphlab.SFrame('image_train_data/')
```

```python
knn_model = graphlab.nearest_neighbors.create(train_data,
                                              features=['deep_features'],
                                              label='id')
```

```python
graphlab.canvas.set_target('ipynb')
cat = train_data[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images(result):
    return train_data.filter_by(result['reference_label'], 'id')
```

```python
cat_neighbors = get_images(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
car = train_data[8:9]
car['image'].show()
```

```python
get_images(knn_model.query(car))['image'].show()
```

```python
find_neighbors = lambda i: get_images(knn_model.query(train_data[i:i+1]))['image'].show()
```
